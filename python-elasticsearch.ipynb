```python
import requests
# Elasticsearch server details
base_url = "http://localhost:9200"  
index_name = "discussions_chat"

# Index settings and mapping
index_settings = {
    "settings": {
        "number_of_shards": 1,
        "number_of_replicas": 0
    },
    "mappings": {
        "properties": {
            'discussions': {"type": "keyword"},
        }
    }
}

# Create the index
response = requests.put(f"{base_url}/{index_name}", json=index_settings)

# Check the response
if response.status_code == 200:
    print(f"Index '{index_name}' created successfully.")
else:
    print(f"Error creating index: {response.content}")
```
